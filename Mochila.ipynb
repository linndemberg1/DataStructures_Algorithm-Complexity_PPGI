```python
# ============================================
# PROBLEMA DA MOCHILA INTEIRA
# PROGRAMAÇÃO DINÂMICA
# ============================================

# --------------------------------------------
# LEITURA DOS DADOS
# --------------------------------------------

n, M = map(int, input("Digite n e M: ").split())

pesos = []
valores = []

print("\nDigite peso e valor de cada item:")

for _ in range(n):
    p, v = map(int, input().split())

    pesos.append(p)
    valores.append(v)

# --------------------------------------------
# TABELA DA PROGRAMAÇÃO DINÂMICA
# --------------------------------------------

dp = [[0 for _ in range(M + 1)] for _ in range(n + 1)]

# --------------------------------------------
# PREENCHIMENTO DA TABELA
# --------------------------------------------

for i in range(1, n + 1):

    for capacidade in range(M + 1):

        # NÃO pega o item
        dp[i][capacidade] = dp[i - 1][capacidade]

        # PEGA o item se couber
        if pesos[i - 1] <= capacidade:

            dp[i][capacidade] = max(
                dp[i][capacidade],
                valores[i - 1] +
                dp[i - 1][capacidade - pesos[i - 1]]
            )

# --------------------------------------------
# VALOR MÁXIMO
# --------------------------------------------

print("\nValor máximo:", dp[n][M])

# --------------------------------------------
# RECUPERAÇÃO DOS ITENS ESCOLHIDOS
# --------------------------------------------

capacidade = M

itens_escolhidos = []

for i in range(n, 0, -1):

    if dp[i][capacidade] != dp[i - 1][capacidade]:

        itens_escolhidos.append(i - 1)

        capacidade -= pesos[i - 1]

# --------------------------------------------
# RESULTADO
# --------------------------------------------

itens_escolhidos.reverse()

print("Itens escolhidos:", itens_escolhidos)

print("\nDetalhes dos itens:")

peso_total = 0

for item in itens_escolhidos:

    print(
        f"Item {item} -> "
        f"Peso = {pesos[item]}, "
        f"Valor = {valores[item]}"
    )

    peso_total += pesos[item]

print("\nPeso total:", peso_total)
```
