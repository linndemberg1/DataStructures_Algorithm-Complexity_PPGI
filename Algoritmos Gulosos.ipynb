## Atividade 1 — Árvore de Espalhamento Mínimo

```python
# ============================================
# ATIVIDADE 1
# KRUSKAL E PRIM
# Árvore de Espalhamento Mínimo
# ============================================

import heapq

# --------------------------------------------
# LEITURA DO GRAFO
# --------------------------------------------

def ler_grafo():
    n = int(input("Digite o número de vértices: "))

    matriz = [[0 for _ in range(n)] for _ in range(n)]

    print("\nDigite os pesos do triângulo superior:")

    for i in range(n - 1):
        valores = list(map(int, input().split()))

        for j, peso in enumerate(valores):
            coluna = i + j + 1

            matriz[i][coluna] = peso
            matriz[coluna][i] = peso

    return matriz


# --------------------------------------------
# KRUSKAL
# --------------------------------------------

class UnionFind:
    def __init__(self, n):
        self.pai = list(range(n))

    def encontrar(self, x):
        if self.pai[x] != x:
            self.pai[x] = self.encontrar(self.pai[x])
        return self.pai[x]

    def unir(self, a, b):
        raizA = self.encontrar(a)
        raizB = self.encontrar(b)

        if raizA != raizB:
            self.pai[raizB] = raizA
            return True

        return False


def kruskal(matriz):
    n = len(matriz)

    arestas = []

    for i in range(n):
        for j in range(i + 1, n):
            arestas.append((matriz[i][j], i, j))

    arestas.sort()

    uf = UnionFind(n)

    mst = []
    custo_total = 0

    for peso, u, v in arestas:
        if uf.unir(u, v):
            mst.append((u, v, peso))
            custo_total += peso

    return mst, custo_total


# --------------------------------------------
# PRIM
# --------------------------------------------

def prim(matriz):
    n = len(matriz)

    visitado = [False] * n

    fila = [(0, 0, -1)]

    mst = []
    custo_total = 0

    while fila:
        peso, u, pai = heapq.heappop(fila)

        if visitado[u]:
            continue

        visitado[u] = True
        custo_total += peso

        if pai != -1:
            mst.append((pai, u, peso))

        for v in range(n):
            if not visitado[v] and matriz[u][v] != 0:
                heapq.heappush(fila, (matriz[u][v], v, u))

    return mst, custo_total


# ============================================
# EXECUÇÃO
# ============================================

matriz = ler_grafo()

print("\n========== KRUSKAL ==========")

mst_kruskal, custo_kruskal = kruskal(matriz)

for u, v, peso in mst_kruskal:
    print(f"{u} -- {v}  peso = {peso}")

print("Custo Total:", custo_kruskal)


print("\n========== PRIM ==========")

mst_prim, custo_prim = prim(matriz)

for u, v, peso in mst_prim:
    print(f"{u} -- {v}  peso = {peso}")

print("Custo Total:", custo_prim)
```

## Atividade 2 — Caminho Mínimo

```python
# ============================================
# ATIVIDADE 2
# DIJKSTRA
# Caminho Mínimo
# ============================================

import heapq

# --------------------------------------------
# LEITURA DO GRAFO
# --------------------------------------------

def ler_grafo():
    n = int(input("Digite o número de vértices: "))

    matriz = [[0 for _ in range(n)] for _ in range(n)]

    print("\nDigite os pesos do triângulo superior:")

    for i in range(n - 1):
        valores = list(map(int, input().split()))

        for j, peso in enumerate(valores):
            coluna = i + j + 1

            matriz[i][coluna] = peso
            matriz[coluna][i] = peso

    return matriz


# --------------------------------------------
# DIJKSTRA
# --------------------------------------------

def dijkstra(matriz, origem=0):
    n = len(matriz)

    dist = [float('inf')] * n
    anterior = [-1] * n

    dist[origem] = 0

    fila = [(0, origem)]

    while fila:
        dist_atual, u = heapq.heappop(fila)

        if dist_atual > dist[u]:
            continue

        for v in range(n):

            peso = matriz[u][v]

            if peso != 0:

                nova_dist = dist[u] + peso

                if nova_dist < dist[v]:
                    dist[v] = nova_dist
                    anterior[v] = u

                    heapq.heappush(fila, (nova_dist, v))

    return dist, anterior


# --------------------------------------------
# RECONSTRUIR CAMINHO
# --------------------------------------------

def reconstruir_caminho(anterior, destino):
    caminho = []

    atual = destino

    while atual != -1:
        caminho.append(atual)
        atual = anterior[atual]

    caminho.reverse()

    return caminho


# ============================================
# EXECUÇÃO
# ============================================

matriz = ler_grafo()

distancias, anterior = dijkstra(matriz, origem=0)

destino = len(matriz) - 1

caminho = reconstruir_caminho(anterior, destino)

print("\n========== DIJKSTRA ==========")

print(f"Menor distância de 0 até {destino}:",
      distancias[destino])

print("Caminho:", caminho)
```
