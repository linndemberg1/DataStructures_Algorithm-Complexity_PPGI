## Implementação dos algoritmos

```python
# MergeSort
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    meio = len(arr) // 2
    esquerda = merge_sort(arr[:meio])
    direita = merge_sort(arr[meio:])

    return merge(esquerda, direita)


def merge(esq, dir):
    resultado = []
    i = j = 0

    while i < len(esq) and j < len(dir):
        if esq[i] < dir[j]:
            resultado.append(esq[i])
            i += 1
        else:
            resultado.append(dir[j])
            j += 1

    resultado.extend(esq[i:])
    resultado.extend(dir[j:])

    return resultado


# QuickSort
def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivo = arr[len(arr) // 2]

    menores = [x for x in arr if x < pivo]
    iguais = [x for x in arr if x == pivo]
    maiores = [x for x in arr if x > pivo]

    return quick_sort(menores) + iguais + quick_sort(maiores)
```

## Execução

```python
def executar(algoritmo, lista):
    if algoritmo == "merge":
        return merge_sort(lista)
    elif algoritmo == "quick":
        return quick_sort(lista)
    else:
        raise ValueError("Algoritmo inválido")


# Exemplo de execução
entrada = [5, 3, 8, 1]

print("Entrada:", entrada)
print("MergeSort:", executar("merge", entrada))
print("QuickSort:", executar("quick", entrada))
```

## Testes

```python
def testar():
    assert merge_sort([3, 1, 2]) == [1, 2, 3]
    assert merge_sort([]) == []
    assert merge_sort([5]) == [5]

    assert quick_sort([3, 1, 2]) == [1, 2, 3]
    assert quick_sort([]) == []
    assert quick_sort([5]) == [5]

    print("Todos os testes passaram!")


testar()
```

## Comparação de desempenho

```python
import random
import time

lista = random.sample(range(10000), 1000)

# MergeSort
inicio = time.time()
merge_sort(lista)
tempo_merge = time.time() - inicio

# QuickSort
inicio = time.time()
quick_sort(lista)
tempo_quick = time.time() - inicio

print(f"Tempo MergeSort: {tempo_merge:.6f} s")
print(f"Tempo QuickSort: {tempo_quick:.6f} s")
```
